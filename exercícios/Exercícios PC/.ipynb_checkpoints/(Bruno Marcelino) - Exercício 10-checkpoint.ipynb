# Exercício 1

```python
def exercicio_1(data):
    meses = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho", "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"]
    elementos = data.split("/")
    elementos[1] = meses[int(elementos[1])-1]
    string = "{0[0]} de {0[1]} de {0[2]}".format(elementos)
    return string
```

# Exercício 2 (desisti, olha o return)

```python
def exercicio_2(texto):
    resultado = []
    
    # espaços em branco
    espacos = len(texto.split(" "))-1
    resultado.append(espacos)
    
    # número de vogais maiúsculas e minúsculas
    vogais = "aeiou"
    vogais_maiusculas = 0
    for i in texto:
        if i in vogais.upper():
            vogais_maiusculas += 1
    
    vogais_minusculas = 0
    for i in texto:
        if i in vogais:
            vogais_minusculas += 1
            
    resultado.append((vogais_maiusculas, vogais_minusculas))
    
    # numero de consoantes maiusculas e minusculas
    consoantes = "bcdfghjklmnpqrstvwxyzç"
    consoantes_maiusculas = 0
    for i in texto:
        if i in consoantes.upper():
            consoantes_maiusculas += 1
    
    consoantes_minusculas = 0
    for i in texto:
        if i in consoantes:
            consoantes_minusculas += 1
            
    resultado.append((consoantes_maiusculas, consoantes_minusculas))
    
    # numero de sinais de pontuação
    sinais = ["?", "!", ".", ",", ":", ";", "\n"]
    num_int, num_excl, num_pontos, num_virgulas, num_dois_pontos, num_pont_virg, num_espaco = 0, 0, 0, 0, 0, 0, 0
    
    for i in texto:
        if i in sinais[0]:
            num_int += 1
    for i in texto:
        if i in sinais[1]:
            num_excl += 1
    for i in texto:
        if i in sinais[2]:
            num_pontos += 1
    for i in texto:
        if i in sinais[3]:
            num_virgulas += 1
    for i in texto:
        if i in sinais[4]:
            num_dois_pontos += 1
    for i in texto:
        if i in sinais[5]:
            num_pont_virg += 1
    for i in texto:
        if i in sinais[6]:
            num_espaco += 1
            
    resultado.append((num_int, num_excl, num_pontos, num_pontos, num_virgulas, num_dois_pontos, num_pont_virg, num_espaco))
    
    return "Não tenho paciência nem tempo para fazer isso."
```

# Exercício 3

```python
def exercicio_3(cpf):
    import re 
    padrao = "\d\d\d\.\d\d\d.\d\d\d-\d\d"
    match = re.match(padrao, cpf)
    
    if match:
        digitos = "".join(re.findall("\d", cpf))
        
        # validação do digito 1
        seq = list(range(10,1,-1))
        soma = 0
        for i in range(len(digitos)-2):
            soma += int(digitos[i]) * seq[i]
        val_1 = 11 - (soma % 11)
        if val_1 < 10:
            digito_1 = val_1
        else:
            digito_1 = 0
        
        # validação do digito 2 
        seq = list(range(11,1,-1))
        soma = 0
        for i in range(len(digitos)-1):
            soma += int(digitos[i]) * seq[i]
        val_2 = 11 - (soma % 11)
        if val_2 < 10:
            digito_2 = val_2
        else:
            digito_2 = 0
            
        if digitos[9] == str(digito_1) and digitos[10] == str(digito_2):
            return True
        else:
            return False
    else: 
        return None
```

# Exercício 4

```python
def exercicio_4(n):
    numeros = ["zero", "um", "dois", "tres", "quatro", "cinco", "seis", "sete", "oito", "nove"]
    mais_de_10 = ["dez", "onze", "doze", "treze", "quatorze", "quinze", "dezesseis", "dezessete", "dezoito", "dezenove"]
    dezenas = ["vinte", "trinta", "quarenta", "cinquenta", "sessenta", "setenta", "oitenta", "noventa"]
    
    if n < 0:
        return None
    elif n == 100:
        return "cem"
    elif 0 <= n <= 9:
        return numeros[n]
    elif 10 <= n <= 19:
        return mais_de_10[n-10]
    else:
        num_split = list(str(n))
        if int(num_split[1]) != 0:
            dezena = dezenas[int(num_split[0])-2]
            unidade = numeros[int(num_split[1])]
            return "{} e {}".format(dezena, unidade)
        else:
            return dezenas[int(num_split[0])-2]
```

# Exercício 5

```python
def exercicio_5_1():
    tabela = ""
    for i in range(2,10):
        colunas = "" 
        for j in range(1,10):
            string = "{0} x {1} = {2}\t".format(i, j, i*j)
            colunas += string
        tabela += colunas[:-1] + "\n"
        
    return tabela[:-1]
```

```python
def exercicio_5_2():
    tabela = ""
    for i in range(1,21):
        colunas = ""
        for j in range(1, 5):
            string = f"{i**j}".rjust(7)
            colunas += string
        tabela += colunas + "\n"
    
    return tabela[:-1]
```

```python
def exercicio_5_3(n):
    linhas = [[1], [1,1]]
    for i in range(n-1):
        linha_ant = linhas[-1]
        
        linha_nova = [1]
        for j in range(i+1):
            num = linha_ant[j] + linha_ant[j+1]
            linha_nova.append(num)
        linha_nova.append(1)   
        
        linhas.append(linha_nova)
    
    string = ""
    for linha in linhas:
        linha_nova = ""
        for num in linha:
            linha_nova += str(num) + "\t\t"
        linha_nova = linha_nova[:-2]
        
        if len(linha) != n+1:
            num_tabs_antes = n - len(linha) # tab = /t
            linha_nova = ("\t"*(num_tabs_antes+1)) + linha_nova + ("\t"*num_tabs_antes)

            string += linha_nova + "\n"
        else:
            num_tabs_antes = n - len(linha) # tab = /t
            linha_nova = ("\t"*(num_tabs_antes)) + linha_nova + ("\t"*num_tabs_antes)

            string += linha_nova + "\n"
        
    return string[:-1]
```
